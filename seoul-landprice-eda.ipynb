```python
data= pd.read_csv("../input/seoul-landprice-and-retail/.csv")
data= data.drop(['Unnamed: 0'],axis=1)

data.head()
```

```python
data['평균소형금액']=data['소형 거래금액']/data['소형 거래건수']
data['평균중소형금액']=data['중소형 거래금액']/data['중소형 거래건수']
data['평균중대형금액']=data['중대형 거래금액']/data['중대형 거래건수']
data['평균대형금액']=data['대형 거래금액']/data['대형 거래건수']
data['유입']=data['전입']-data['전출']

data
```

```python
data_col= data[['행정동','인구','유입','제산세','평균소형금액','평균중소형금액','평균중대형금액','평균대형금액','총개소','임대료','임대면적','보증금','공시지가']]
data_col.columns=['region','population','flow','income','avgsmall','avgmedsmall','avgmedbig','avgbig','totalretail','rent','depo','rentsize','landprice']
data_col=data_col.fillna(0)
data_col
```

# k means clustering

```python
data_col
```

```python
from sklearn import preprocessing


processed_data_col = data_col.copy()
scaler = preprocessing.MinMaxScaler()

processed_data_col[['population', 'flow','income','totalretail']] = scaler.fit_transform(processed_data_col[['population', 'flow','income','totalretail']])
processed_data_col
```

```python
from sklearn.cluster import KMeans

for i in range(1, 7):
       # 클러스터 생성
    estimator = KMeans(i)
    ids = estimator.fit_predict(processed_data_col[['population', 'flow','income','totalretail']])
       # 2행 3열을 가진 서브플롯 추가 (인덱스 = i)
    plt.figure(figsize=(12,8))
    plt.subplot(3, 2, i)
        # 서브플롯의 라벨링
    
    plt.title("K value = {}".format(i))
    plt.xlabel('income')
    plt.ylabel('population')
        # 클러스터링 그리기
    plt.scatter(processed_data_col['income'], processed_data_col['population'], c=ids)  
    plt.legend(fontsize=14)
plt.show()
```

```python
estimator = KMeans(3)
ids = estimator.fit_predict(processed_data_col[['population', 'flow','income','totalretail']])
ids
```

```python
processed_data_col['label']= ids
processed_data_col
```

```python
m= folium.Map(
location=[37.566345, 126.977893],
tiles= 'Stamen Toner',)

folium.Choropleth(
geo_data= geo_json,
name='choropleth',
data=processed_data_col,
columns=['region','label'],
key_on='feature.properties.name',
fill_color='BuPu',
fill_opacity=0.7,
line_opacity=0.2).add_to(m)

m
```

# Mice algorithm

```python
df_imp = data_col[["region",'population','flow','income','totalretail','landprice','rent']].replace(0, np.NaN)
df_impute= df_imp[['population','flow','income','totalretail','landprice','rent']]
df_impute
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imp_mean = IterativeImputer(random_state=0)
fit= imp_mean.fit_transform(df_impute)
```

```python
df_fit= pd.DataFrame(fit)
df_fit.columns=['population','flow','income','totalretail','landprice','rent']
df_fit['region']=df_imp['region']
df_fit
```

```python
df_fit['rent-price ratio']= df_fit['rent']/df_fit['landprice']*100000
```

```python
import folium
geo_json='https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_submunicipalities_geo_simple.json'

m= folium.Map(
location=[37.566345, 126.977893],
tiles= 'Stamen Toner',)

folium.Choropleth(
geo_data= geo_json,
name='choropleth',
data=df_fit,
columns=['region','rent-price ratio'],
key_on='feature.properties.name',
fill_color='BuGn',
fill_opacity=0.7,
line_opacity=0.2).add_to(m)

m
```
